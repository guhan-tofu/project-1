#Find best selling month

```python
#create a month column
all_data["Month"]=all_data["Order Date"].str[0:2]
```

```python
all_data.head()
```

Changing Price dtype to float

```python
all_data["Price Each"]=all_data["Price Each"].astype('float64')
```

```python
all_data.drop(all_data[all_data["Month"]=="Or"].index,inplace=True)
```

```python
all_data["Price Each"]=all_data["Price Each"].astype('float64')
```

```python
all_data["Quantity Ordered"]=pd.to_numeric(all_data["Quantity Ordered"])
```

Cleaning dataset

```python
all_data[all_data.isna().any(axis=1)]
#we can see that there are a  lot of rows with Nan in all columns
```

```python
all_data.dropna(how="all")
#how can be "any"(any one column with NaN) or "all"(all columns in a row with NaN)
```

```python
all_data['Sales']= all_data["Quantity Ordered"] * all_data["Price Each"]
```

```python
sale=all_data.groupby("Month").agg({"Sales":"sum"}).sort_values("Sales",ascending=False)
```

# Therefore December is the Month with Highest Revenue

```python
sns.barplot(x=sale.index,y="Sales",data=sale,palette="viridis")
```

#What time should you advertise for most sales

```python
all_data["Order Date"]= pd.to_datetime(all_data["Order Date"])
# only disadvantage is datetime is larger data than string
```

```python
all_data.head()
```

```python
all_data["Hour"]=all_data["Order Date"].dt.hour
```

```python
time=all_data.groupby("Hour").agg({"Sales":"sum"})
```

#Therefore 7pm and 11am is the peak hours to run Ads

```python
fig,ax=plt.subplots(figsize=(10,6))
ax.plot(time.index,time["Sales"],c="green")
plt.xticks(time.index[:])
ax.set_xlabel("Hour")
ax.set_ylabel("Sales")
ax.axvline(19,ls="--")
ax.axvline(11,ls="--")
ax.text(11.2,2370000,"<--These are the peak hours -->");
```

#What products are most often sold together

```python
# Gives all rows with duplicate Order ID therefore now we have a dataset with more than one item both under same Order ID only
df=all_data[all_data["Order ID"].duplicated(keep=False)] #keep must be "first" or "last" (meaning which duplicate to keep) passing False keeps both
```

```python
df
```

```python
df["Grouped"]=df.groupby("Order ID")['Product'].transform(lambda x : ",".join(x))
```

```python
df
```

```python
df=df.drop_duplicates(subset=["Order ID"])
```

```python
df
```

```python
result=df.groupby("Grouped").agg({"Grouped":'count'})
```

```python
df_new = result.rename(columns={'Grouped': 'Count'})
```

```python
df_new=df_new.sort_values("Count",ascending=False).iloc[0:5]
```

```python
df_new
```

#Top 5 products sold together most often

```python
sns.barplot(x="Count",y=df_new.index,data=df_new)
```
